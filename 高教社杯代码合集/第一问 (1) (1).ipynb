```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# 定义自变量X和因变量y
X = data_clean[['检测孕周', '孕妇BMI']]  # 自变量
y = data_clean['Y染色体浓度']  # 因变量

# 多项式特征生成（选择最高多项式次数为2）
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# 拟合多项式回归模型
poly_model = LinearRegression()
poly_model.fit(X_poly, y)

# 输出回归系数
print("Polynomial regression coefficients:", poly_model.coef_)

y_pred = poly_model.predict(X_poly)
residuals = y - y_pred
mse = np.mean(residuals**2)
se = np.sqrt(mse / len(X))  # 简单的标准误差估算
print("Standard Error:", se)
```

```python
from sklearn.linear_model import Ridge, Lasso
from sklearn.model_selection import cross_val_score
import numpy as np

# 定义自变量和因变量
X = data_clean[['检测孕周', '孕妇BMI']]
y = data_clean['Y染色体浓度']

# Ridge回归（岭回归）
ridge_model = Ridge(alpha=1.0)  # alpha是正则化参数，控制正则化强度
ridge_score = cross_val_score(ridge_model, X, y, cv=5, scoring='neg_mean_squared_error')  # 使用交叉验证
print(f"Ridge Model MSE (cross-validation): {-ridge_score.mean()}")  # 输出平均均方误差

# Lasso回归（套索回归）
lasso_model = Lasso(alpha=0.01)  # alpha是正则化参数
lasso_score = cross_val_score(lasso_model, X, y, cv=5, scoring='neg_mean_squared_error')  # 使用交叉验证
print(f"Lasso Model MSE (cross-validation): {-lasso_score.mean()}")  # 输出平均均方误差

# 选择最佳模型（如果你认为 Ridge 回归更好，可以选择它，或者 Lasso 回归）
ridge_model.fit(X, y)  # 拟合岭回归模型
lasso_model.fit(X, y)  # 拟合套索回归模型

# 输出回归系数
print(f"Ridge Coefficients: {ridge_model.coef_}")
print(f"Lasso Coefficients: {lasso_model.coef_}")
ridge_pred = ridge_model.predict(X)
residuals = y - ridge_pred
mse_ridge = np.mean(residuals**2)
se_ridge = np.sqrt(mse_ridge / len(X))  # 标准误差
print("Ridge Standard Error:", se_ridge)
lasso_pred = lasso_model.predict(X)
residuals = y - lasso_pred
mse_lasso = np.mean(residuals**2)
se_lasso = np.sqrt(mse_lasso / len(X))  # 标准误差
print("Lasso Standard Error:", se_lasso)
```

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import Ridge, Lasso, LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
from scipy import stats

# 准备数据
X = data_clean[['检测孕周', '孕妇BMI']]  # 自变量
y = data_clean['Y染色体浓度']  # 因变量

# 添加常数项（截距）
X = sm.add_constant(X)

# 1. 多项式回归
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# 拟合多项式回归模型
poly_model = LinearRegression()
poly_model.fit(X_poly, y)

# 获取回归系数
poly_coefs = poly_model.coef_

# 计算残差
y_pred_poly = poly_model.predict(X_poly)
residuals_poly = y - y_pred_poly

# 计算均方误差（MSE）和标准误差（SE）
mse_poly = np.mean(residuals_poly**2)
se_poly = np.sqrt(mse_poly / len(X))  # 估计标准误差

# 计算t值和p值
t_values_poly = poly_coefs / se_poly
p_values_poly = [2 * (1 - stats.t.cdf(np.abs(t), df=len(X) - 1)) for t in t_values_poly]

t_values_poly = np.round(t_values_poly, 2)
p_values_poly = np.round(p_values_poly, 2)

# 显示结果
print("Polynomial Regression t-values:", t_values_poly)
print("Polynomial Regression p-values:", p_values_poly)

# 2. Ridge回归（岭回归）
ridge_model = Ridge(alpha=1.0)
ridge_model.fit(X, y)

# 获取岭回归的系数
ridge_coefs = ridge_model.coef_

# 计算残差
y_pred_ridge = ridge_model.predict(X)
residuals_ridge = y - y_pred_ridge

# 计算均方误差（MSE）和标准误差（SE）
mse_ridge = np.mean(residuals_ridge**2)
se_ridge = np.sqrt(mse_ridge / len(X))

# 计算t值和p值
t_values_ridge = ridge_coefs / se_ridge
p_values_ridge = [2 * (1 - stats.t.cdf(np.abs(t), df=len(X) - 1)) for t in t_values_ridge]

t_values_ridge = np.round(t_values_ridge, 2)
p_values_ridge = np.round(p_values_ridge, 2)

# 显示结果
print("Ridge Regression t-values:", t_values_ridge)
print("Ridge Regression p-values:", p_values_ridge)

# 3. Lasso回归（套索回归）
lasso_model = Lasso(alpha=0.1)
lasso_model.fit(X, y)

# 获取Lasso回归的系数
lasso_coefs = lasso_model.coef_

# 计算残差
y_pred_lasso = lasso_model.predict(X)
residuals_lasso = y - y_pred_lasso

# 计算均方误差（MSE）和标准误差（SE）
mse_lasso = np.mean(residuals_lasso**2)
se_lasso = np.sqrt(mse_lasso / len(X))

# 计算t值和p值
t_values_lasso = lasso_coefs / se_lasso
p_values_lasso = [2 * (1 - stats.t.cdf(np.abs(t), df=len(X) - 1)) for t in t_values_lasso]

t_values_lasso = np.round(t_values_lasso, 2)
p_values_lasso = np.round(p_values_lasso, 2)

# 显示结果
print("Lasso Regression t-values:", t_values_lasso)
print("Lasso Regression p-values:", p_values_lasso)
```

```python
from sklearn.linear_model import Ridge, Lasso
from sklearn.model_selection import GridSearchCV

# 为 Ridge 和 Lasso 设置不同的 alpha 值进行网格搜索
ridge_params = {'alpha': [0.1, 1, 10, 100]}
lasso_params = {'alpha': [0.1, 1, 10, 100]}

ridge_model = Ridge()
lasso_model = Lasso()

# 网格搜索优化 Ridge 模型
ridge_search = GridSearchCV(ridge_model, ridge_params, cv=5)
ridge_search.fit(X, y)
print(f"Best alpha for Ridge: {ridge_search.best_params_}")

# 网格搜索优化 Lasso 模型
lasso_search = GridSearchCV(lasso_model, lasso_params, cv=5)
lasso_search.fit(X, y)
print(f"Best alpha for Lasso: {lasso_search.best_params_}")
```

```python
best_model = ridge_search.best_estimator_  # 假设选择 Ridge
y_pred = best_model.predict(X)

# 输出预测结果
print("Predicted values:", y_pred)
```
