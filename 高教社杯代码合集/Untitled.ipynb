```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
male_data = pd.read_excel(r'男胎.xlsx')
female_data = pd.read_excel(r'女胎.xlsx')
combined_data = pd.concat([male_data, female_data], ignore_index=True)
```

```python
combined_data
```

```python
import matplotlib.pyplot as plt

# 男胎和女胎的数量
male_count = 1082
female_count = 605

labels = ['男胎', '女胎']
sizes = [male_count, female_count]
plt.rcParams['font.sans-serif'] = ['SimHei']  # 设置字体为黑体
plt.rcParams['axes.unicode_minus'] = False  # 用来正常显示负号
# 绘制饼状图
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightcoral'])

# 设置标题
plt.title('男女胎儿的性别分布', fontsize=16)

# 显示图像
plt.show()
```

```python
import seaborn as sns

# 绘制孕周分布的柱状图
plt.figure(figsize=(10, 6))
sns.histplot(combined_data['检测孕周'], bins=15, kde=False, color='skyblue', stat='count')

# 设置标题和标签
plt.title('Distribution of gestational weeks', fontsize=16)
plt.xlabel('Detecting gestational weeks', fontsize=12)
plt.ylabel('Number', fontsize=12)

# 显示图像
plt.show()
```

```python
import matplotlib.pyplot as plt
x = male_data[['检测孕周']]  # 自变量
y = male_data['Y染色体浓度']  # 因变量
# 假设你已经有了 'Y染色体浓度' 和 '检测孕周' 这两列数据，分别为 y_chromosome_concentration 和 detection_week
plt.figure(figsize=(8,6))

# 绘制散点图
plt.scatter(x, y, color='blue', alpha=0.6)

# 添加标题和标签
plt.title('Y染色体浓度与检测孕周的散点图')
plt.xlabel('检测孕周')
plt.ylabel('Y染色体浓度')

# 显示图形
plt.show()
```
