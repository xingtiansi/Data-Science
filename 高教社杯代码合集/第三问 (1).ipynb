```python
import statsmodels.api as sm
import numpy as np

# 假设 cleaned_data 已经准备好并包含所需的变量（身高、体重、年龄、BMI 和 检测孕周）

# 自变量：身高、体重、年龄和 BMI
X = cleaned_data[['年龄', '孕妇BMI']]

# 因变量：检测孕周
y = cleaned_data['检测孕周']

# 添加常数项（截距）
X = sm.add_constant(X)

# 拟合 OLS 回归模型
model = sm.OLS(y, X).fit()

# 输出回归结果
print(model.summary())
```

```python
cleaned_data['BMI_squared'] = cleaned_data['孕妇BMI'] ** 2
cleaned_data['BMI_age_interaction'] = cleaned_data['孕妇BMI'] * cleaned_data['年龄']
cleaned_data['age_squared'] = cleaned_data['年龄'] ** 2
# 自变量：身高、体重、年龄、BMI、BMI的二次项、BMI与年龄的交互项
X = cleaned_data[['年龄', '孕妇BMI', 'age_squared']]
y = cleaned_data['检测孕周']

# 添加常数项（截距）
X = sm.add_constant(X)

# 拟合 OLS 模型
model = sm.OLS(y, X).fit()

# 输出回归结果
print(model.summary())
```

```python
from linearmodels.panel import PanelOLS

# 假设数据中有 '孕妇ID' 和 '时间' 变量，用于标识面板数据
cleaned_data['time'] = cleaned_data['时间']
cleaned_data['id'] = cleaned_data['序号']

# 设置面板数据结构
panel_data = cleaned_data.set_index(['id', 'time'])

# 自变量和因变量
X = panel_data[['身高', '体重', '年龄', 'BMI']]
y = panel_data['检测孕周']

# 固定效应回归模型
fe_model = PanelOLS(y, X, entity_effects=True).fit()

# 输出回归结果
print(fe_model.summary())
```

```python
from linearmodels.panel import RandomEffects
import pandas as pd

# 假设数据已准备好，包含了 '孕妇ID' 和 '时间' 作为面板数据的标识
cleaned_data['time'] = cleaned_data['时间']
cleaned_data['id'] = cleaned_data['孕妇ID']

# 设置面板数据结构，id 是个体标识，time 是时间标识
panel_data = cleaned_data.set_index(['id', 'time'])

# 自变量和因变量
X = panel_data[['身高', '体重', '年龄', 'BMI']]
y = panel_data['检测孕周']

# 随机效应回归模型
re_model = RandomEffects(y, X).fit()

# 输出回归结果
print(re_model.summary())
```

```python
np.random.seed(42)
error = np.random.normal(0, 1, len(y))  # 模拟误差，均值为0，标准差为1

# 为预测值添加误差
y_pred_with_error = model.predict(X) + error

# 计算误差模型下的 MSE
mse_with_error = np.mean((y_pred_with_error - y) ** 2)
print(f"MSE with error: {mse_with_error}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 绘制拟合值与实际值的对比图
plt.figure(figsize=(10, 6))
sns.scatterplot(x=y, y=y_pred_with_error, color='blue')
plt.title('Fitted Values with Error vs Actual Values', fontsize=16)
plt.xlabel('Actual Values', fontsize=12)
plt.ylabel('Fitted Values with Error', fontsize=12)
plt.grid(True)
plt.show()

# 绘制残差图
plt.figure(figsize=(10, 6))
sns.scatterplot(x=y_pred_with_error, y=y - y_pred_with_error, color='red')
plt.axhline(0, color='black', linestyle='--')
plt.title('Residuals vs Fitted Values with Error', fontsize=16)
plt.xlabel('Fitted Values with Error', fontsize=12)
plt.ylabel('Residuals', fontsize=12)
plt.grid(True)
plt.show()
```
