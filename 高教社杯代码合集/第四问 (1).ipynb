```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
from sklearn.preprocessing import StandardScaler

# 假设 cleaned_data 是包含所有特征和标签的 DataFrame
# 特征包括 '21号非整倍体', '18号非整倍体', '13号非整倍体', 'X染色体Z值', 'GC含量', '读段数', '相关比例', 'BMI'
# 目标变量是 '异常'，表示是否异常

# 提取特征和目标变量
X = female_data[['21号染色体的Z值', '18号染色体的Z值', '13号染色体的Z值', 
                        'X染色体的Z值', 'GC含量', '唯一比对的读段数', '被过滤掉读段数的比例', '孕妇BMI']]
y = female_data['胎儿是否健康']  # 0=正常, 1=异常

# 数据标准化
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 划分训练集和测试集
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# 构建随机森林模型
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# 预测并评估模型
y_pred = rf_model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
from sklearn.model_selection import cross_val_score

# 使用交叉验证评估模型
cv_scores = cross_val_score(rf_model, X_scaled, y, cv=5)  # 5折交叉验证
print("Cross-validation scores:", cv_scores)
print("Average cross-validation score:", cv_scores.mean())
```

```python
from sklearn.model_selection import GridSearchCV

# 定义参数网格
param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [10, 20, 30],
    'min_samples_split': [2, 5, 10]
}

# 使用网格搜索进行超参数调优
grid_search = GridSearchCV(RandomForestClassifier(), param_grid, cv=5)
grid_search.fit(X_train, y_train)

# 输出最佳参数和最佳得分
print("Best Parameters:", grid_search.best_params_)
print("Best Cross-validation Score:", grid_search.best_score_)
```

```python
best_rf_model = RandomForestClassifier(
    max_depth=grid_search.best_params_['max_depth'],
    min_samples_split=grid_search.best_params_['min_samples_split'],
    n_estimators=grid_search.best_params_['n_estimators'],
    random_state=42
)

# 训练模型
best_rf_model.fit(X_train, y_train)

# 预测测试集
y_pred = best_rf_model.predict(X_test)

# 输出评估指标
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
import pandas as pd
import numpy as np

# 读取数据
data = pd.read_excel("female_cleaned_data.xlsx")

# 定义Z值计算函数
def calculate_z(X, mu, sigma):
    """
    计算Z值
    :param X: 样本中的目标染色体的相对计数比例
    :param mu: 对照组中该染色体计数比例的均值
    :param sigma: 对照组中该比例的标准差
    :return: 计算得到的Z值
    """
    return (X - mu) / sigma

# 设置阈值
z_threshold = 3  # Z值阈值，大于3或小于-3认为是异常

# 假设的正常对照组的均值和标准差，这些值应根据实际数据填写
mu_21 = -1.226006e-01  # 假设21号染色体正常浓度的均值
sigma_21 = 1.112380e+00  # 21号染色体正常浓度的标准差

mu_18 = 7.863986e-01  # 假设18号染色体正常浓度的均值
sigma_18 = 4.230630e-03  # 18号染色体正常浓度的标准差

mu_13 = 3.790528e-01  # 假设13号染色体正常浓度的均值
sigma_13 = 1.334025e+00  # 13号染色体正常浓度的标准差

mu_x = 4.598381e-01  # 假设X染色体正常浓度的均值
sigma_x = 1.317251e+00  # X染色体正常浓度的标准差

# 计算Z值
data['Z_21'] = calculate_z(data['21号染色体的Z值'], mu_21, sigma_21)
data['Z_18'] = calculate_z(data['18号染色体的Z值'], mu_18, sigma_18)
data['Z_13'] = calculate_z(data['13号染色体的Z值'], mu_13, sigma_13)
data['Z_X'] = calculate_z(data['X染色体的Z值'], mu_x, sigma_x)

# 判断是否异常的函数
def judge_abnormal(row):
    """
    判断女胎是否异常
    :param row: 数据行
    :return: 异常标记
    """
    abnormal_flag = False
    if abs(row['Z_21']) > z_threshold or abs(row['Z_18']) > z_threshold or abs(row['Z_13']) > z_threshold:
        abnormal_flag = True
    if abs(row['Z_X']) > z_threshold:
        abnormal_flag = True
    return abnormal_flag

# 应用判断函数
data['Abnormal'] = data.apply(judge_abnormal, axis=1)

# GC含量判断（假设正常范围为40%-60%）
def check_gc_content(row):
    """
    判断GC含量是否在正常范围内
    :param row: 数据行
    :return: 是否在正常范围内
    """
    return 0.40 <= row['GC含量'] <= 0.60

# 应用GC含量判断
data['GC_Normal'] = data.apply(check_gc_content, axis=1)

# 读段数判断（假设正常范围为50%~100%）
def check_read_count(row):
    """
    判断读段数是否符合标准
    :param row: 数据行
    :return: 是否符合标准
    """
    return 0.50 <= row['在参考基因组上比对的比例'] <= 1.00

# 应用读段数判断
data['Read_Count_Normal'] = data.apply(check_read_count, axis=1)

# 综合判断是否有效
data['Final_Judgment'] = data['Abnormal'] & data['GC_Normal'] & data['Read_Count_Normal']

# 输出判断结果
data[['孕妇代码', 'Final_Judgment']].to_excel("female_abnormality_judgment.xlsx", index=False)

print("判断完成，结果已输出到 'female_abnormality_judgment.xlsx'")
```
