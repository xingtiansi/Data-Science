```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
male_data = pd.read_excel(r'male_cleaned_data.xlsx')
```

```python
import re

def extract_week(x):

    if '+' in x:
   
        week_number = int(re.findall(r'\d+', x)[0])  # 提取 'w' 前面的数字
        extra_days = int(re.findall(r'\d+', x.split('+')[1])[0])  # 提取 '+' 后面的数字
        return week_number + extra_days / 7
    else:
 
        return int(re.findall(r'\d+', x)[0])


male_data['检测孕周'] = male_data['检测孕周'].apply(extract_week)
```

```python
male_data.describe().T
```

```python
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf

# 假设数据已经加载到 `data` 中，包含列 'GW', 'BMI', 'Y染色体浓度', '孕妇ID', '检测次数'
# 示例数据加载（请用实际数据替代）
# data = pd.read_csv("your_data.csv")

# 创建一个新的变量，表示孕妇ID和检测次数的组合
male_data['woman_id_time'] = male_data['孕妇代码'].astype(str) + "_" + male_data['检测抽血次数'].astype(str)
male_data['孕妇BMI_square'] = male_data['孕妇BMI'] ** 2
male_data['BMI_检测孕周_interaction'] = male_data['孕妇BMI'] * male_data['年龄']
male_data['检测孕周_square'] = male_data['检测孕周'] ** 2
male_data['检测孕周_cubic'] = male_data['检测孕周'] ** 3
# 设置自变量 (GW, BMI) 和因变量 (Y染色体浓度)
X = male_data[['检测孕周', '孕妇BMI', '检测孕周_square','检测孕周_cubic', '孕妇BMI_square','BMI_检测孕周_interaction','woman_id_time']]
y = male_data['Y染色体浓度']

# 加入常数项 (截距)
X = sm.add_constant(X)

# 使用混合效应模型，'woman_id_time' 为随机效应
model = smf.mixedlm("Y染色体浓度 ~ 检测孕周 + 孕妇BMI + 检测孕周_square + 检测孕周_cubic + 孕妇BMI_square + BMI_检测孕周_interaction", male_data, groups=male_data["woman_id_time"]).fit()

# 输出结果
print(model.summary())
```

```python
from sklearn.metrics import mean_squared_error
y_pred = model.fittedvalues

# 获取实际值
y_actual = male_data['Y染色体浓度']

# 计算MSE
mse = mean_squared_error(y_actual, y_pred)

print("Mixed Effects Model MSE:", mse)
```
