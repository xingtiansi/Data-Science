```python
bins = [25, 28, 31, 34, 37, 40]  # 设定 BMI 区间
labels = ['[25, 28)', '[28, 31)', '[31, 34)', '[34, 37)', '[37, 40)']
cleaned_data['BMI_group'] = pd.cut(cleaned_data['孕妇BMI'], bins=bins, labels=labels, right=False)

# 查看 BMI 分组
print(cleaned_data[['孕妇BMI', 'BMI_group']])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# 假设数据已经加载到 cleaned_data DataFrame 中
# 使用 apply() 方法进行 BMI 分组
def bmi_group(bmi):
    if 25 <= bmi < 28:
        return '[25, 28)'
    elif 28 <= bmi < 31:
        return '[28, 31)'
    elif 31 <= bmi < 34:
        return '[31, 34)'
    elif 34 <= bmi < 37:
        return '[34, 37)'
    elif 37 <= bmi < 40:
        return '[37, 40)'
    else:
        return 'Out of range'

# 对 BMI 列进行分组
cleaned_data['BMI_group'] = cleaned_data['孕妇BMI'].apply(bmi_group)

# 统计每个 BMI 组的数量
group_counts = cleaned_data['BMI_group'].value_counts()

# 绘制柱状图
plt.figure(figsize=(10, 6))
sns.barplot(x=group_counts.index, y=group_counts.values, palette='viridis')

# 设置图表标题和标签
plt.title('BMI Group Distribution', fontsize=16)
plt.xlabel('BMI Groups', fontsize=12)
plt.ylabel('Number of Individuals', fontsize=12)
plt.xticks(rotation=45)  # 使X轴标签倾斜，避免重叠

# 显示图像
plt.show()
```

```python
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# 假设 cleaned_data 已经准备好，并且有 BMI 和 检测孕周数据

# 按照 BMI 分组进行回归分析
for group in cleaned_data['BMI_group'].unique():
    # 获取当前组的数据
    group_data = cleaned_data[cleaned_data['BMI_group'] == group]
    
    # 自变量和因变量
    X = group_data[['孕妇BMI']]  # 这里的 'BMI' 列作为自变量
    y = group_data['检测孕周']  # '检测孕周' 作为因变量
    
    # 添加常数项（截距）
    X = sm.add_constant(X)
    
    # 拟合 OLS 模型
    ols_model = sm.OLS(y, X).fit()
    
    # 获取拟合值和残差
    y_pred = ols_model.predict(X)
    residuals = y - y_pred
    
    # 绘制拟合值 vs 实际值图
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x=y, y=y_pred, color='blue')
    plt.title(f'Fitted Values vs Actual Values for BMI Group {group}', fontsize=16)
    plt.xlabel('Actual Values', fontsize=12)
    plt.ylabel('Fitted Values', fontsize=12)
    plt.grid(True)
    plt.show()

    # 绘制残差图
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x=y_pred, y=residuals, color='red')
    plt.axhline(0, color='black', linestyle='--')
    plt.title(f'Residuals vs Fitted Values for BMI Group {group}', fontsize=16)
    plt.xlabel('Fitted Values', fontsize=12)
    plt.ylabel('Residuals', fontsize=12)
    plt.grid(True)
    plt.show()

    # 绘制QQ图
    sm.qqplot(residuals, line ='45', fit=True)
    plt.title(f'QQ Plot of Residuals for BMI Group {group}', fontsize=16)
    plt.show()

    # 输出回归结果
    print(f"Regression Results for BMI Group {group}:")
    print(ols_model.summary())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
import numpy as np

# 假设我们已经有 cleaned_data，并且数据包含 BMI 和 检测孕周
X = cleaned_data[['孕妇BMI', '检测孕周']].dropna()  # 去除缺失值

# 执行KMeans聚类分析
kmeans = KMeans(n_clusters=3, random_state=42)
cleaned_data['cluster'] = kmeans.fit_predict(X)

# 获取聚类中心
cluster_centers = kmeans.cluster_centers_

# 可视化聚类结果
plt.figure(figsize=(10, 6))

# 绘制散点图，每个点的颜色代表其所属的簇
sns.scatterplot(x='孕妇BMI', y='检测孕周', hue='cluster', data=cleaned_data, palette='viridis', s=100, alpha=0.7)

# 绘制聚类中心（每个簇的最优检测时点）
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], s=200, c='red', marker='X', label='Cluster Centers')

# 设置图像标题和标签
plt.title('KMeans Clustering: BMI and Best NIPT Detection Week', fontsize=16)
plt.xlabel('BMI', fontsize=12)
plt.ylabel('Best NIPT Detection Week', fontsize=12)
plt.legend()

# 显示图像
plt.grid(True)
plt.show()
```
