```python
df.columns = [x.strip() for x in col] #除去列名里的空格，万一有的话
```

```python
df.duplicated() #返回布尔型数据，告诉重复值的位置
```

```python
df[df.duplicated()] #这个数据是清洗过的 所以查看重复值的时候发现是没有的 数据是干净的
```

```python
df.duplicated().sum() #所以是0个
```

```python
df.drop_duplicates() #删去重复值
```

```python
df.shape[0] #这里因为没有重复值 所以个数没有减少
```

```python
df.index = range(df.shape[0])   #因为删去一些数据之后 我们要更新他的index比如说第5个数被删了 那么这个index空出来了 后面要补上去 这一步相当于更新index
```

```python
df.describe().T
```

```python
sta = (df['WKSESL']-df['WKSESL'].mean())/df['WKSESL'].std() #超出三倍标准差 找到具体的数据 需要单独考虑一下 是不是有问题 建模的时候通常删掉
```

```python
sta.abs()>3
```

```python
df[sta.abs()>3]
```

```python
delindex = pd.concat([df[sta.abs()>3]]).index
```

```python
df.drop(delindex, inplace = True)
```

```python
df.shape #异常值已经删除了
```

```python
df.index = range(df.shape[0])
```

```python
df[df.P1FSTAMP.isnull()] #查看具体缺失值的行
```

```python
df['P1FSTAMP'].fillna(round(df['P1FSTAMP'].mean(),0), inplace=True) #均值填补就行 如果是数值类数据的话 其他的话到时候看到再解决
```
